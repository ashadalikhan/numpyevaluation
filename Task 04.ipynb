```python
import numpy as np
```

```python
from matplotlib.pyplot import imread
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.ndimage.filters import convolve
```

```python
Asad = np.array([[-1, 0, 1], 
               [-1, 0, 1], 
               [-1, 0, 1] ])
```

```python
img = imread("123.jpg")
```

```python
channels = []
for channel in range(3):
    res = convolve(img[:,:,channel], Asad)
    channels.append(res)

img = np.dstack((channels[0], 
                 channels[1], 
                 channels[2]))
```

```python
plt.imshow(img)
plt.show()
```
